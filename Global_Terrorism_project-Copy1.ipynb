# 5.Rename Columns

```python
df.rename(columns={'iyear':'Year','imonth':'Month','iday':"day",'gname':'Group','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','latitude':'latitude',
    'longitude':'longitude','summary':'summary','attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'kill',
     'nwound':'Wound'},inplace=True)
```

```python
df = df[['Year','Month','day','Country','State','Region','City','latitude','longitude',"Attacktype",'kill',
               'Wound','target1','summary','Group','Targettype','Weapon','motive']]
```

# 6.Handling Null values

```python
df['kill']=df['kill'].fillna(0)
df['Wound']=df['Wound'].fillna(0)
df['Total_Casualty']=df['kill']+df['Wound']
df.head()
```

# 7.Performing Some Data Visualizations:

```python
import matplotlib.pyplot as plt

years = df['Year'].unique()
years_count = df['Year'].value_counts(dropna=False).sort_index()

plt.figure(figsize=(18, 10))
plt.plot(years_count.index, years_count.values, marker='o', color='blue')
plt.xticks(rotation=60)
plt.xlabel('Attacking Year', fontsize=20)
plt.ylabel('Number of Attacks Each Year', fontsize=20)
plt.title('Attacks In Years', fontsize=17)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

year = df['Year'].unique()
years_count = df['Year'].value_counts(dropna=False).sort_index()

plt.figure(figsize=(15, 7))
sns.barplot(x=year, y=years_count, palette="tab10")
plt.xticks(rotation=60, fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel('Attacking Year', fontsize=10)
plt.ylabel('Number of Attacks Each Year', fontsize=14)
plt.title('Attacks In Years', fontsize=18)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

cross_table = pd.crosstab(df['Year'], df['Region'])
colors = ['blue', 'green', 'orange', 'red', 'purple']  # Specify colors for each region

ax = cross_table.plot.area(stacked=False, figsize=(20, 10), color=colors)

ax.set_title('Terrorist Activities By Region In Each Year', fontsize=25)
ax.set_ylabel('Number of Attacks', fontsize=20)
ax.set_xlabel('Year', fontsize=20)

plt.show()
```

```python
import matplotlib.pyplot as plt

data = df[['Year', 'kill']].groupby(['Year']).sum().reset_index()

fig, ax4 = plt.subplots(figsize=(15, 6))
ax4.bar(data['Year'], data['kill'], alpha=0.7)

plt.xticks(rotation=50)
plt.title("People Died Due To Attack", fontsize=12)
plt.ylabel("Number of killed people", fontsize=12)
plt.xlabel('Year', fontsize=12)

ax4.spines["top"].set_visible(False)
ax4.spines["right"].set_visible(False)

plt.show()
```

```python
attack_counts = df['Attacktype'].value_counts()

fig, ax = plt.subplots(figsize=(14, 5))
ax.bar(attack_counts.index, attack_counts, color='cyan',width=0.6)

plt.xticks(rotation=55)
plt.xlabel('Attack Type', fontsize=10)
plt.ylabel('Number of Attacks', fontsize=10)
plt.title('Number of Attacks by Attack Type', fontsize=15)

plt.show()
```

```python
group_kill_sum = df[['Group', 'kill']].groupby('Group').sum().drop('Unknown').sort_values('kill', ascending=False).head(10)

fig, ax = plt.subplots(figsize=(17, 7))
ax.bar(group_kill_sum.index, group_kill_sum['kill'], color='yellow')

plt.title('Top 10 Terrorist Group Attacks', fontsize=15)
plt.xticks(rotation=55)
plt.xlabel('Terrorist Group Name', fontsize=15)
plt.ylabel('Number of Killed People', fontsize=15)

plt.show()
```
